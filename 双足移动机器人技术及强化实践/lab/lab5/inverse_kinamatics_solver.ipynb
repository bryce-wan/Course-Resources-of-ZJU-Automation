```python
import sympy as sym
import numpy as np

# Define the symbols
theta1, theta2, theta3, theta4, theta5, theta6 = sym.symbols('theta1 theta2 theta3 theta4 theta5 theta6')
d1, d2, d4, d5, d6 = sym.symbols('d1 d2 d4 d5 d6')
alpha1, alpha4, alpha5 = sym.symbols('alpha1 alpha4 alpha5')
a3, a2 = sym.symbols('a3 a2')
T06, T01, T12, T23, T34, T45, T56 = sym.symbols('T06 T01 T12 T23 T34 T45 T56')

theta_lists = [theta1, theta2, theta3, theta4, theta5, theta6]
d_lists = [d1, d2, 0, d4, d5, d6]
alpha_lists = [alpha1, 0, 0, alpha4, alpha5, 0]
a_lists = [0, a2, a3, 0, 0, 0]
T_lists = [T01, T12, T23, T34, T45, T56]
```

```python
def DH2T(a, alpha, d, theta):
    return sym.Matrix([[sym.cos(theta), -sym.cos(alpha) * sym.sin(theta), sym.sin(alpha) * sym.sin(theta), a * sym.cos(theta)],
                    [sym.sin(theta), sym.cos(alpha) * sym.cos(theta), -
                     sym.sin(alpha) * sym.cos(theta), a * sym.sin(theta)],
                    [0, sym.sin(alpha), sym.cos(alpha), d],
                    [0, 0, 0, 1]])
```

```python
T06 = (np.eye(4)).astype(np.int16)
for i in range(6):
    T_lists[i] = DH2T(a_lists[i], alpha_lists[i], d_lists[i], theta_lists[i])
    T06 = T06 @ T_lists[i]
    print(T_lists[i])
    print('')
print(T06)
```
